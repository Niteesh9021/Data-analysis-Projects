# Correlation Test

```python
correlation=data.corr()
correlation.iloc[1].reset_index()
```

* The correlation results shows that the GDP Growth Rate can be defined 54% by Unemployment rate, 5% by FDI, 2% by Inflation rate.

* Unemployment is in the strongest correlation with GDP growth (-0.537) and the weakest correlation is Inflation with GDP growth (-0.025). 

* Multiple correlation coefficient of the dependent variable (GDP growth rate) and independent variables (FDI as % of GDP, Unemployment, Inflation) is 0.600, meaning that their connection is strong. 

* Determination coefficient is 0.360, meaning that 36 % of the variability of the dependent variable (GDP growth rate) could be explained by the effect of the independent variables (FDI as % of GDP, Unemployment, Inflation) and it means that there are various other variables which affects the dependent variable.

### CONCLUSION

* Economic theory and practice do not know any unique set of indicators that affect GDP. The very complexity of this phenomenon relatives any attempt to establish an unchanged list of indicators. With the purpose of quantification of relation between macroeconomic factors and GDP, we have used multiple linear regression model. 

* The analysis showed that GDP with as much as 36% can be explained by the influence of independent variables: FDI, Unemployment, Inflation.

* Out of all the analysed independent variables, unemployment has major effect on GDP, followed by FDI and Inflation.
